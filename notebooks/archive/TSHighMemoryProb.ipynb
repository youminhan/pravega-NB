```python
import pandas
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
from praveganb.idracdemo import *
%matplotlib inline
```

# Calculate the probability of Running on >20% Memory Bandwidth utilization 

```python
usage = 20
```

```python
dfs = pandas.DataFrame()
read_events = calculate_grsf(usage=usage,
                              rack='eaglemonk',
                              metricId='MemoryUsage',
                              reportid='CUPS',
                              scope='idracdemo',
                              stream='idracdata',
                              gateway='pravega-grpc-gateway:80'
                             )
for df in read_events:
    clear_output(wait=True)
    dfs = dfs.append(df)
    f, ax = plt.subplots(sharey=True)
    dfs['grsf'].plot(ax=ax, figsize=(10,5), grid=True)
    ax.legend(loc='center left')
    plt.title("TS Calculation of {0} Memory usage".format(usage))
    ax.set_ylabel("Probability % ")
    plt.show()
```
